# 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.



```python
# Below Function will calculate median of meduim income column.
def calculate_median(data):
    sorted_data = sorted(data)  # Sorted function will sort data but will not modify it.
    data_length = len(sorted_data) # len function it will get the length of the sorted_data.

    # If-else condition will check length of data is even or odd.
    if data_length % 2 == 0:
        middle_index_1 = data_length // 2 - 1
        middle_index_2 = data_length // 2
        median = (sorted_data[middle_index_1] + sorted_data[middle_index_2]) / 2
    # If the length is even, take the average of the two middle elements
    else:
        middle_index = data_length // 2
        median = sorted_data[middle_index]
    # If the length is odd, the median is the middle element
    return median

# It will load the dataset.
df = pd.read_csv("/content/California Housing Dataset.csv")
# This line will run and calculate the median using user defined function.
median_income_median = calculate_median(df['median_income'])

print("Median Income Median:", median_income_median)
```

# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# Importing required libraries.
import pandas as pd

# It will load our dataset.
data = pd.read_csv("/content/California Housing Dataset.csv")

# Below function will segregate bedrooms according to size.
def categorize_bedroom_size(value):
    if value <= 10:
        return 'Small'
    elif 11 <= value < 1000:
        return 'Medium'
    else:
        return 'Large'

# It will create separate column as 'total_bedroom_size' by using apply() method.
data['total_bedroom_size'] = data['total_bedrooms'].apply(categorize_bedroom_size)

# It will show the original dataset with the new column as 'total_bedroom_size'.
print("Dataset with 'total_bedroom_size' column:")
print(data)
```
